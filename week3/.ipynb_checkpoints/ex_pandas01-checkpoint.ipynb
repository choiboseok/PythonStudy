### 3. **데이터 순회하기, 인덱스와 열로 액세스**
  - 'iterrows()' DataFrame의 각 행을 (index, Serice) 쌍으로 반환
  - 'iloc[0]' 행과 열의 위치(정수 인덱스)를 사용하여 데이터에 액세스
  - 'loc[0, 'column_name'] 행과 열의 라벨을 사용하여 데이터에 액세스

```python
for idx, row in df.iterrows():
    print(idx, row)
```

```python
print(df.iloc[0]) # 첫 번째 행 가져오기
```

```python
print(df.iloc[:, 0]) # 첫 번째 열 가져오기
```

```python
df.loc['2010-06-29', 'Close'] # 인덱스, 컬럼
```
