## 1. 선형 회귀 : 예측하고 수정하기(먼저 선 긋고 수정하기)
- 선형회귀는 연속된 값을 예측할때 사용하며 (데이터에 따라 선형, 비선형, 다항) 회귀가 있음.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# x:height, y:weight
df = pd.read_csv("../dataset/heights.csv")
df.head()
x = np.array(df['height'].round())
y = np.array(df['height'].round())
plt.scatter(x, y)
```

```python
# 임의의 기울기 : a 와 y 절편 : b 값을 초기화
a = 0.1
b = 0.1
# 학습틀
lr = 0.0001
# 몇번 학습할지
epochs = 2001
# x 수량
n = len(x)
```

```python
# optimizer(경사하강법)
for i in range(epochs):
    y_pred = a * x + b # 예측값을 구하는 식
    error = y - y_pred # 정답(실제값)과 예측값의 차이를 구함.

    a_diff = (2 / n) * sum(-x * (error)) # 오차 함수를 a로 편미분한 값.
    b_diff = (2 / n) * sum(-(error)) # 오차 함수를 b로 편미분한 값

    a = a - lr * a_diff # 학습률을 곱해 기존의 a값을 업데이트
    b = b - lr * b_diff # ...
    if i % 100 == 0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" %(i, a, b))
```

```python
# 최종 a, b를 대입하여 그래프를 출력
y_pred = a*x+b
plt.scatter(x, y)
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], 'r')
```

```python
# 공부시간 x1, 과외시간 x2, 성적 y
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
```

```python
# y = a1*x1 + a2*x2 + b
a1 = 0
a2 = 0
b = 0 
lr = 0.01
epoch = 2001
n = len(x1)
```

```python
for i in range(epochs):
    y_pred = a1*x1 + a2*x2 + b
    error = y*y_pred
    a1_diff = (2/n) * sum(-x1*(error))
    a2_diff = (2/n) * sum(-x2*(error))
    b_diff = (2/n) * sum(-(error))

    a1 = a1-lr*a1_diff
    a2 = a1-lr*a2_diff
    b = b-lr*b_diff
    if i % 100 == 0:
        print("epoch=%.f, 기울기=%.04f, 기울기2=%.04f, 절편=%.04f" %(i, a1, a2, b))
```
